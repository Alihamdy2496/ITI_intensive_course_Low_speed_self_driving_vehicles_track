```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img1 = cv2.imread('Building.jpg')
img2 = cv2.imread('Road.jpg')

gray1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(25, 10))
plt.imshow(img1[:,:,::-1]);
```

```python
def nothing(x):
    pass

cv2.namedWindow('Hough')

cv2.createTrackbar('th', 'Hough', 12, 500, nothing)
cv2.createTrackbar('minLineLength', 'Hough', 1, 800, nothing)
cv2.createTrackbar('maxLineGap', 'Hough', 6, 800, nothing)

edges1 = cv2.Canny(gray1, 50, 200)

while 1:
    image = img1.copy()
    th = cv2.getTrackbarPos('th','Hough')
    minLineLength = cv2.getTrackbarPos('minLineLength','Hough')
    maxLineGap = cv2.getTrackbarPos('maxLineGap','Hough')

    lines1 = cv2.HoughLinesP(edges1, 1, np.pi/180, th, minLineLength=minLineLength, maxLineGap=maxLineGap )
    if lines1 is not None:
        for i in range(0, len(lines1)):
            x1=lines1[i][0][0]
            y1=lines1[i][0][1]
            x2=lines1[i][0][2]
            y2=lines1[i][0][3]
            cv2.line(image, (x1, y1), (x2, y2), (0, 0, 255), 2)
    
    edgesC1 = cv2.cvtColor(edges1, cv2.COLOR_GRAY2BGR)
    image = cv2.addWeighted(image, 0.9, edgesC1, 0.5, 0)
    cv2.imshow('Hough', image)
    k = cv2.waitKey(1)
    if k == 27:
        plt.imsave("out1.jpg",image[:,:,::-1])
        break
cv2.destroyAllWindows()
```
